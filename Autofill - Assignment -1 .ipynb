```python
import nltk
from nltk import word_tokenize,sent_tokenize
import numpy as np
import string
import re
import tkinter


count=0
probability={}    
n_grams={}
n=3

# Open File.
file= open('test.txt' ,"r", encoding="utf8") 
data=file.read()
file.close()


# Tokens.
#pattern = '^\s [$&+,:;=?@#|<>.^*()%!-]| [\u0600-\u06ff]|[\u0750-\u077f]|[\ufb50-\ufbc1]|[\ufbd3-\ufd3f]|[\ufd50-\ufd8f]|[\ufd92-\ufdc7]|[\ufe70-\ufefc]|[\uFDF0-\uFDFD]/'
pattern = '[^\s ء-ي]'
replacement = ''
data=re.sub(pattern, replacement,data)
tokens=nltk.word_tokenize(data)
#print(tokens)


# Calculate probability .
for i in range(len(tokens)-(n)):
    sequence=' '.join(tokens[i:i+n])
    if sequence not in n_grams.keys():
        n_grams[sequence]=1
        count +=1
    else:
        n_grams[sequence]+=1
        count+=1
    probability[sequence]=n_grams[sequence]/count

print(probability.keys())


 
    
# Create Prediction function Using probability .








# GUI .
    
    
    
```
